```python
import urllib.request
import pandas as pd
from astropy.coordinates import SkyCoord
from astroquery.sdss import SDSS
from astropy import units as u

# Load the dataset
data = pd.read_csv('GalaxyZoo1_DR_table2.csv')

# Convert the ra and dec columns to decimal format
coords = SkyCoord(data['RA'], data['DEC'], unit=(u.hourangle, u.deg))
ra_decimal = coords.ra.deg
dec_decimal = coords.dec.deg

```

```python
import os
def fetch_image(ra, dec, size=1024, output_file='galaxy.jpg'):
    url = f'http://skyservice.pha.jhu.edu/DR12/ImgCutout/getjpeg.aspx?ra={ra}&dec={dec}&scale=0.15&width=256&height=256'
    urllib.request.urlretrieve(url, output_file)

# Create folders for each category
if not os.path.exists("spiral"):
    os.mkdir("spiral")
if not os.path.exists("elliptical"):
    os.mkdir("elliptical")
    
    
# Iterate through the rows in the dataset
for i, row in data.iterrows():
    # Check the value of the 'spiral' and 'elliptical' columns
    if row['SPIRAL'] == 1 or row['ELLIPTICAL'] == 1:
        # Fetch and download the image
        fetch_image(ra_decimal[i], dec_decimal[i], output_file=f'galaxy_{i}.jpg')

        # Check the value of the 'spiral' and 'elliptical' columns again
        if row['SPIRAL'] == 1:
            os.rename(f'galaxy_{i}.jpg', f'spiral/galaxy_{i}.jpg')
        elif row['ELLIPTICAL'] == 1:
            os.rename(f'galaxy_{i}.jpg', f'elliptical/galaxy_{i}.jpg')
```

```python
from skimage import io
from skimage.filters import unsharp_mask

# Location of the original images
original_folder = "spiral"

# Create a new folder for the sharpened images
sharpened_folder = "galaxy_images_sharpened"
if not os.path.exists(sharpened_folder):
    os.mkdir(sharpened_folder)

# Apply unsharp masking to all images in the folder
for filename in os.listdir(original_folder):
    if filename.endswith(".jpg"):
        image = io.imread(f"{original_folder}/{filename}")
        sharpened_image = unsharp_mask(image, radius=5, amount=1)
        io.imsave(f"{sharpened_folder}/{filename}", sharpened_image)
```
